```python
# Import dependencies
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 

# import file directory from computer
file_dir = "C:/Users/16are/Desktop/UNC_BootCamp/Homework_Github/Module_14/bikesharing/"

# The CitiBike CSV File
bike_link = f'{file_dir}201908-citibike-tripdata.csv'

# Have python read the csv file
bike_data = pd.read_csv(bike_link, low_memory=False)

# Read csv file into a dataframe
bike_df = pd.DataFrame(bike_data)

# Display dataframe
bike_df.sample(10)
```

```python
# 2. Check the datatypes of your columns. 
bike_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
bike_df['tripduration'] = pd.to_datetime(bike_df['tripduration'], unit='s')

# Display dataframe column
bike_df['tripduration']
```
